```python
import pandas as pd
import pymysql
import datetime
```

```python
df = pd.DataFrame()
for y in range(2000, 2018):
        df_tmp = pd.read_csv("EUR_USD_4H/"+ str(y) +"_4H.csv", encoding="utf-8")
        df = pd.concat([df, df_tmp], axis = 0, ignore_index=True)
df
```

```python
df["DateTime"] = pd.to_datetime(df["DateTime"], format='%Y-%m-%d %H:%M:%S')
```

```python
from sqlalchemy import create_engine
engine = create_engine("mysql+pymysql://root:123456@db/iii_project", echo=False)
```

```python
df.to_sql('eur_usd_4hr', con=engine, if_exists="append", index=False)
engine.execute("SELECT * FROM eur_usd_4hr").fetchall()
```
